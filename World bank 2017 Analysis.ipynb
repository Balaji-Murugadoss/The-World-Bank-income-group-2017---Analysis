#### **Q1**

You have a rich uncle who has peacefully passed away in his sleep. In his will you have been named to receive USD100,000. Assuming you have good lawyers and accountants, you are able to legally avoid all taxes associated with this windfall and the funds are immediately deposited to an investment account in the Bahamas that earns 1.0% interest that is compounded every month. Write a Python program using a while loop that will calculate how many months will be needed for the balance of the account to reach $1,500,000.

The goal is to retire and just live on the interest for the rest of your life. Output a table showing the balance in each month until you reach your goal. Be sure to label the columns clearly and format the balance in each row as $xxx,xxx.xx

```python
# Initial deposit
principal = 100000

# Monthly interest rate
monthly_interest_rate = 0.01

# Target balance
target_balance = 1500000

# Initialize a counter for months
months = 0

# Print the header for the table
print("Month\tBalance")
print("-----\t-------")

# Use a while loop to calculate the number of months and print the table
while principal < target_balance:
    # Calculate the interest earned in the current month
    interest = principal * monthly_interest_rate

    # Add the interest to the principal
    principal += interest

    # Increment the month counter
    months += 1

    # Print the month and balance in a formatted way
    print(f"{months}\t${principal:,.2f}")

# Print the final message
print("It will take", months, "months to reach $1,500,000.")
```

**Q5**

Write a **for** loop that will print out a list of all the geos in americas that are in the High Income category according to the World Bank income group 2017. Also produce a list comprehension syntax to accomplish the same thing as the **for** loop.

```python
import pandas as pd

# Load the data from Excel
xls = pd.ExcelFile('Data Geographies - v1 - by Gapminder.xlsx')
countries_df = pd.read_excel(xls, 'List of countries etc')

# Assuming 'List of countries etc' contains columns 'Country', 'Income Group', and 'Region'
# Define the region of interest
americas_region = 'americas'

# Initialize an empty list to store high income countries in the Americas
high_income_americas = []

# Iterate over each row in the DataFrame
for index, row in countries_df.iterrows():
    if americas_region in row['four_regions'] and row['World bank income group 2017'] == 'High income':
        high_income_americas.append(row['geo'])

# Print the list of high income countries in the Americas
for country in high_income_americas:
    print(country)
```
